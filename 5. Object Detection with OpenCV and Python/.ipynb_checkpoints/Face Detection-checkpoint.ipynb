```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np
```

```python
nadia = cv2.imread("Nadia_Murad.jpg",0)
denis = cv2.imread("Denis_Mukwege.jpg",0)
solvay = cv2.imread("solvay_conference.jpg",0)
```

```python
plt.imshow(nadia,cmap='gray')
```

```python
plt.imshow(denis,cmap='gray')
```

```python
plt.imshow(solvay,cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result = detect_face(denis)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result = detect_face(solvay)
plt.imshow(result,cmap='gray')
```

```python
result = detect_face(nadia)
plt.imshow(result,cmap='gray')
```

```python
def adj_detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result = adj_detect_face(solvay)
plt.imshow(result,cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    face_img = img.copy()
    eyes_rects = eye_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    for (x,y,w,h) in eyes_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result = detect_eyes(nadia)
plt.imshow(result,cmap='gray')
```

```python
# White around the pupils is not distinct enough to detect Denis' eyes here!
result = detect_eyes(denis)
plt.imshow(result,cmap='gray')
```

```python
# Face Detection with video
cap = cv2.VideoCapture(0)
while True:
    ret,frame = cap.read()
    frame = detect_face(frame)
    cv2.imshow('Video Face Detect',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
