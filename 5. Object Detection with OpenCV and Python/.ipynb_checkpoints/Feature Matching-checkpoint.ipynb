```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np
```

```python
def display_img(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
# target image
reeses = cv2.imread("reeses_puffs.png",0)
display_img(reeses)
```

```python
# larger secondary image
cereals = cv2.imread("many_cereals.jpg",0)
display_img(cereals)
```

# Brute Force Detection with ORB Descriptors

```python
# creating detection object to detect features
orb = cv2.ORB_create()
```

```python
kp1, des1 = orb.detectAndCompute(reeses,None)
kp2, des2 = orb.detectAndCompute(cereals,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches = bf.match(des1,des2)
```

```python
len(matches)
```

```python
# less distance = better match & more distance = worst match
# Matches are sorted by distance attribute
matches = sorted(matches,key=lambda x:x.distance)
```

```python
reeses_matches = cv2.drawMatches(reeses,kp1,cereals,kp2,matches[:25],None,flags=2)
```

```python
display_img(reeses_matches)
```

# Brute-Force Matching with SIFT Descriptors and Ratio Test

```python
# creating detection object to detect features
sift = cv2.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(reeses,None)
kp2, des2 = sift.detectAndCompute(cereals,None)
```

```python
bf = cv2.BFMatcher()
```

```python
matches = bf.knnMatch(des1,des2,k=2) # Here k=2 means 1st and 2nd best matches
matches
```

```python
good = []
#Ratio Test
for match1,match2 in matches:
    # if match1 distance is less than 75% of match2 distance
    # Then descriptor was a good match, Lets keep it.
    if match1.distance < 0.75*match2.distance:
        good.append([match1])
```

```python
good
```

```python
len(matches)
```

```python
len(good)
```

```python
sift_matches = cv2.drawMatchesKnn(reeses,kp1,cereals,kp2,good,None,flags=2)
```

```python
display_img(sift_matches)
```

# FLANN based Matcher

```python
sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(reeses,None)
kp2, des2 = sift.detectAndCompute(cereals,None)
```

```python
# FLANN parameters
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks=50)  
```

```python
flann = cv2.FlannBasedMatcher(index_params,search_params)
```

```python
matches = flann.knnMatch(des1,des2,k=2)
```

```python
good = []
#Ratio Test
for match1,match2 in matches:
    # if match1 distance is less than 75% of match2 distance
    # Then descriptor was a good match, Lets keep it.
    if match1.distance < 0.75*match2.distance:
        good.append([match1])
```

```python
flann_matches = cv2.drawMatchesKnn(reeses,kp1,cereals,kp2,good,None,flags=0)
display_img(flann_matches)
```
