```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np
```

```python
img = cv2.imread("sammy_face.jpg")
plt.imshow(img) # No need to convert image into rgb as we are performing edge detecion only
```

```python
edges = cv2.Canny(image=img,threshold1=127,threshold2=127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img,threshold1=0,threshold2=255)
plt.imshow(edges)
```

```python
med_val = np.median(img)
med_val
```

```python
# Lower threshold to either 0 or 70% of the median value whichever is greater
lower = int(max(0,0.7*med_val))
# Upper threshold to either 130% of the median value or the max 255, whichever is smaller
upper = int(min(255,1.3*med_val))
```

```python
lower
```

```python
upper
```

```python
edges = cv2.Canny(image=img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
# For better result we will blurred the image first, and then we will apply Canny Edge Detection
blurred_img = cv2.blur(img,ksize=(5,5))
```

```python
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
# For more better result 
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```
