```python
# resizing image
new_img = cv2.resize(fix_img,(1000, 400)) #(width, height)
```

```python
plt.imshow(new_img)
```

```python
# resizing image with ratios
w_ratio = 0.8
h_ratio = 0.2 # 20% of original height
new_img2 = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img2)
```

```python
new_img2.shape
```

```python
# flipping image
new_img3 = cv2.flip(fix_img, 0) #horizontal
```

```python
plt.imshow(new_img3)
```

```python
new_img3 = cv2.flip(fix_img, 1) #vertical
```

```python
plt.imshow(new_img3)
```

```python
new_img3 = cv2.flip(fix_img, -1) #both
```
