```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img = cv2.imread("00-puppy.jpg")
```

```python
plt.imshow(img)
```

```python
# Converting BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
```

```python
plt.imshow(img)
```
