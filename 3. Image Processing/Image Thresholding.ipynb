```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img = cv2.imread("rainbow.jpg",0) #0 means reading image in grayscale
```

```python
plt.imshow(img,cmap='gray')
```

```python
#255/2=threshold value
#any value below 127 in image will be shifted to 0, any value above 127 in image will be shifted to max value which is 255
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY) 
```

```python
ret
```

```python
plt.imshow(thresh1,cmap='gray') #all the values are in 0 or 255 (binary image)
```

```python
#any value below 127 in image will be shifted to 255, any value above 127 in image will be shifted to 0
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV) 
```

```python
plt.imshow(thresh1,cmap='gray') #inverse of the binary result
```

```python
#any value above 127 in image will be shifted to threshold(127), any value below 127 in image will remain as it is
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC) 
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
#any value above 127 in image will remain as it is, any value below 127 in image will be shifted to 0
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO) 
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
#inverse of threshold TOZERO
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV) 
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
img = cv2.imread("crossword.jpg",0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
ret,th1 = cv2.threshold(img,180,255,cv2.THRESH_BINARY) 
```

```python
show_pic(th1) 
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_pic(blended)
```
